```python
%%bash
source ~/.bashrc
conda activate blobtools
for i in cnat dcyl ssid; do
    cd /storage/group/ibb3/default/acer_cnat_dcyl_ssid_assembly/pacbio_assembly/${i}_hifiasm_meta_2/blobtools_out/
    awk '$15~/Suessiales/' \
        out.blobDB.table.txt \
        | cut -f1 | sort \
        > sym_contigs.txt
    blobtools seqfilter -i ../${i}_asm.fa -l sym_contigs.txt -o sym_contigs
done
```

```python
%%bash
cd /storage/group/ibb3/default/acer_cnat_dcyl_ssid_assembly/
wget --no-check-certificate https://symportal.org/static/resources/SymPortal_unique_DIVs.tar.gz
tar -xzvf ./SymPortal_unique_DIVs.tar.gz
```

```python
%%bash
source ~/.bashrc
conda activate assembly
for i in cnat dcyl ssid; do
    cd /storage/group/ibb3/default/acer_cnat_dcyl_ssid_assembly/pacbio_assembly/${i}_hifiasm_meta_2/blobtools_out/
    blastn -query sym_contigs.${i}_asm.filtered.fna \
        -subject /storage/group/ibb3/default/acer_cnat_dcyl_ssid_assembly/SymPortal_unique_DIVs.fasta \
        -outfmt 6 -out results.txt
done
```
